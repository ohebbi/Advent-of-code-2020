```python
#2a
import pandas as pd
input2 = pd.read_csv("adventofcode2.txt", sep= " ", header=None)
totalsum = 0
for row in input2.values:
    rowsum = 0
    for letter in row[2]:
        if letter == row[1][0]:
            rowsum += 1
    tmp = row[0].split("-")
    if rowsum >= int(tmp[0]) and int(tmp[1])>=rowsum:
        totalsum += 1
print(totalsum)
```

```python
#2b

totalsum = 0
for row in input2.values[:]:
    rowsum = 0
    tmp = row[0].split("-")
    if row[2][int(tmp[0])-1] == row[1][0]:
        rowsum += 1
    if row[2][int(tmp[1])-1] == row[1][0]:
        rowsum += 1
    
    #print(tmp)
    if rowsum == 1:
        totalsum += 1
print(totalsum)
```

```python
#3a
input2 = pd.read_csv("3a.txt", header=None)

def function3b(right, down):
    
    
    totalTree = 0
    position = 0
    for row_index in range(down,len(input2[0]),down):

        position += right
        
        if position >= len(input2[0][row_index]):
            position -= len(input2[0][row_index])
        #print(row_index)
        if input2[0][row_index][position] == "#":
            totalTree += 1
    
    return totalTree    

print(function3b(3,1))
```
